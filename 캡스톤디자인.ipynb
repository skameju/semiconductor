```python
import pandas as pd
df = pd.read_csv('fab.csv')
df.head() # 처음 5개 행 데이터 확인
```

```python
import matplotlib.pyplot as plt
df_1 = df.set_index('SensorTime')
df_1 = df_1.sort_index(ascending = True)
df_1
```

```python
import numpy as np
ax = df_1.plot(y = ['Sensor0', 'Sensor1', 'Sensor2'], figsize = (15, 6))
total_len = len(df_1.index)
ax.set_xticks(np.arange(0, total_len + 1, 100))
plt.xticks(rotation = 45)
plt.show()
```

```python
data = df_1.reset_index(drop=True)
data # SensorTime 지운 후 data 확인
```

```python
# 공정 합격, 불량 개수 count하고 pie chart로 확인 결과 pass 93.4%, fail 6.6%
count_pass = len(data.loc[data['Pass_Fail'] == -1])
count_fail = len(data.loc[data['Pass_Fail'] == 1])

labels = ['Pass', 'Fail']
values = [count_pass, count_fail]

plt.pie(values, labels = labels, autopct = '%.1f')
plt.show()
```

```python
data.info(verbose=True, null_counts=True)
```

```python
# 컬럼별 결측치 개수 확인
df.isnull().sum()
```

```python
# 결측치 비율 확인
missing_num = data.isnull().sum()/len(data)
missing_num
```

```python
# 결측치 평균 값으로 채우기
data = data.fillna(data.mean())
```

```python
# 결측치 없어진 것을 알 수 있음
missing_num = data.isnull().sum()/len(data)
missing_num
```

```python
data_a = data.corrwith(data['Pass_Fail']).abs() # Pass_Fail과 각 센서들 간의 상관계수의 절댓값 계산
data_Corr = data_a.drop(['Pass_Fail']) # Pass_Fail 행 삭제
data_Corr
```

```python
ax = data_Corr.plot.bar(figsize = (15, 6), color = 'orange')
total_len = len(data_Corr.index)
ax.set_xticks(np.arange(0, total_len + 1, 22))
plt.xlabel('Sensor')
plt.ylabel('Sum(pass fail_abs)')
plt.xticks(rotation = 60)
plt.show() # 상관성 그래프
```

```python
data_Corr_sort = data_Corr.sort_values(ascending = False) #내림차순 정리
data_Corr_sort
```

```python
data_Corr_sort = data_Corr_sort[:40].reset_index() # 상위 40개 확인
data_Corr_sort
```

```python
# 데이터에서 상관 계수가 높은 센서 40개 열만 추출하여
# data 변수에 저장
df = data[data_Corr_sort['index'].tolist()]
df
```

```python
df_x_data = df
df_y_data = data['Pass_Fail']
```

```python
# train, test set 분할
'''
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size = 0.4, random_state = 89)
train_set = train_set.reset_index(drop = True)
test_set = test_set.reset_index(drop = True)

'''

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x_data, df_y_data, train_size = 0.7, test_size = 0.3, shuffle = False)
```

```python
ax = train_x.plot(y = ['Sensor21', 'Sensor28', 'Sensor59'], figsize = (15, 6))
```

# 04-29 수정한 부분

```python
# 정규화

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_minmax_scaled = scaler.fit_transform(train_x)

# test_x도 정규화 해주어야 함
test_x_scaled = scaler.transform(test_x)
```

```python
train_x_scaled = pd.DataFrame(train_minmax_scaled, columns = train_x.columns, index = list(train_x.index.values))
train_x_scaled # 정규화된 train_set!!!!!

# test_x 정규화 함
test_x_scaled = pd.DataFrame(test_x_scaled, columns = test_x.columns, index = list(test_x.index.values))
test_x_scaled
```

```python
ax = train_x_scaled.plot(y = ['Sensor21', 'Sensor28', 'Sensor59'], figsize = (10, 6))
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=0)
x_train_sm, y_train_sm = SMOTE(random_state=0).fit_resample(train_x_scaled, train_y)
```

```python
pd.Series(y_train_sm).value_counts()
```

```python
import sklearn.svm as svm

svm_model = svm.SVC(kernel='rbf', C=8, gamma = 0.1)

svm_model.fit(x_train_sm, y_train_sm)
```
